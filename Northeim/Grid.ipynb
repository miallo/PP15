```python
import random
import numpy as np
from __future__ import print_function
import math
```

```python
def la(x):
    return x*(40000/360)
def lo(x,y):
    return la(x)*math.sin(math.pi/2-math.pi/180*y)

a = 9.961131
b = 51.728028

x0 = lo(a,b)
y0 = la(b)
```

```python
with open( "superTiefe.csv", "r" ) as mergeData:
     allData = [line.strip().split('\t') for line in mergeData]
allData = [[float(first), float(second), float(third)] for first,second,third in allData]
allData = [[1000*(lo(second,first)-x0),1000*(la(first)-y0) , third] for first,second,third in allData]
print(allData)
```

```python
xmin=-400
xmax=1000
ymin=-1600
ymax=200

size=5
radius=25

grid = []
for i in range(0, (xmax-xmin)/size):
    grid.append([])
    for j in range(0, (ymax-ymin)/size):
        grid[i].append([xmin+i*size,ymin+j*size,0.0,0.0])
        #print grid[i][j]
```

```python
def dist(x1,y1,x2,y2):
    return math.sqrt((x1-x2)**2+(y1-y2)**2)
```

```python
for i in range(len(grid)):
    for j in range(len(grid[i])):
        for k in range(len(allData)):
            if(dist(grid[i][j][0],grid[i][j][1],allData[k][0],allData[k][1])<=radius):
                grid[i][j][2]+=allData[k][2]
                grid[i][j][3]+=1.0
```

```python
f1=open('./test.txt', 'w+')
for i in range(len(grid)):
    for j in range(len(grid[i])):
        if(grid[i][j][2]!=0.0):
            grid[i][j][2]/=grid[i][j][3]
        print(grid[i][j][0],grid[i][j][1],grid[i][j][2],file=f1)
    print("",file=f1)
f1.close()
```
